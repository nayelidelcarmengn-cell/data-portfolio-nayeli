# A soup for rents

```python
from bs4 import BeautifulSoup
import requests
import html5lib
import pandas as pd
import csv

#Preparing to starage in a csv file
csv_file=open('rentarenmerida.csv','w')
csv_writer=csv.writer(csv_file)
csv_writer.writerow(['bedroom','bathroom','area','parking','price'])

# Difine function to get a soup

def made_soup(url):
    req=requests.get(url)
    html_file=req.text
    soup=BeautifulSoup(html_file,'lxml')
    return soup

# first page and its soup
url='https://www.vivanuncios.com.mx/s-casas-en-renta/yucatan/v1c1299l1030p1'
soup=made_soup(url)

#Getting trough all pages resulst in a csv_file



url_pages=[]
for result in soup.find_all("a",class_="page-box"):
    try:
        url_p='https://www.vivanuncios.com.mx'+str(result).split('"')[3]
        url_pages.append(url_p)

    except:
          url_pages.append(url)
#Take an eye
url_pages
```

```python
# Extrac the characters and labels

for url_p in url_pages:
    soup=made_soup(url_p)
    for house_rent in soup.find_all('div', class_="tileV2 REAdTileV2 regular listView"):
    #for house_rent in soup.find_all('a', class_="href-link tile-title-text"):
        try:
            bedroom=house_rent.find('div',class_="chiplets-inline-block re-bedroom").text
            bathroom=house_rent.find('div',class_="chiplets-inline-block re-bathroom").text
            area=house_rent.find('div',class_="chiplets-inline-block surface-area").text.split(" ")[0]
            parking=house_rent.find('div',class_="chiplets-inline-block car-parking").text
            price=house_rent.find('span',class_="ad-price").text.split("$")[1].split('\n')[0] #label
        except:
            pass
        csv_writer.writerow([bedroom,bathroom,area,parking,price])
csv_file.close()
```
