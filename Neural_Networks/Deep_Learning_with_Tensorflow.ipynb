# Comparación de modelos

```python
import tensorflow as tf
import pathlib
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
tf.random.set_seed(42)
```

## Preparación de los datos de prueba

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
DIR_BASE='/content/drive/MyDrive/DL_tensorflow/datos/objetos/'
DIR_BASE = pathlib.Path(DIR_BASE)
tam_lote = 32
alto_img = 150
ancho_img = 150
conj_prueba = tf.keras.preprocessing.image_dataset_from_directory( DIR_BASE,
    validation_split=0.2,
    subset = 'validation',
    seed = 53,
    image_size = (alto_img,ancho_img),
    batch_size = tam_lote
)
conj_prueba_bis = tf.keras.preprocessing.image_dataset_from_directory( DIR_BASE,
    validation_split=0.2,
    subset = 'validation',
    seed = 53,
    image_size = (224,224),
    batch_size = tam_lote
)
```

## Carga de los modelos

```python
modelo_diseñado = tf.keras.models.load_model('/content/drive/MyDrive/DL_tensorflow/modelo_disenado')
```

```python
modelo_preentrenado= tf.keras.models.load_model('/content/drive/MyDrive/DL_tensorflow/modelo_preentrenado')
```

```python
modelo_teachable = tf.keras.models.load_model('/content/drive/MyDrive/DL_tensorflow/modelo_teachable/keras_model.h5',compile=False)

modelo_teachable.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate =0.0001),
                     loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

## Resúmenes de los modelos

```python
modelo_diseñado.summary()
```

```python
modelo_preentrenado.summary()
```

```python
modelo_teachable.summary()
```

## Función de pérdida y exactitud

```python
perdida_prueba_1, exactitud_prueba_1 = modelo_diseñado.evaluate(conj_prueba,batch_size = tam_lote, verbose =1)
perdida_prueba_2, exactitud_prueba_2 = modelo_preentrenado.evaluate(conj_prueba,batch_size = tam_lote, verbose =1)
perdida_prueba_3, exactitud_prueba_3 = modelo_teachable.evaluate(conj_prueba_bis,batch_size = tam_lote, verbose =1)
```

**Coclusiones: **

Ninguno de los 3 modelos presenta un nivel de exactitud para ser desplegado para resolver un problema real. Una forma de mejorar el desempeño de los modelos es aumentar el número de imágenes en la base de datos (se entreno usando solo 100 imágenes aprox. para cada clase).

En cuanto los resultados obtenidos hasta el momento, el modelo preentrenado supera a los otro dos, debido a que su función de pérdida es menor y su nivel de exactitud (proporción de imágenes clasificadas correctamente) es mayor.
